```python
df[
    df.isnull().any(axis=1) | 
    df.apply(lambda row: row.astype(str).str.lower().eq("null").any(), axis=1)
]
```

```python
df.head()
```

```python
df["score_improvement"] = df["final_score"] - df["previous_score"]
```

```python
df["performance_band"] = pd.cut(
    df["final_score"],
    bins=[0, 60, 75, 90, 100],
    labels=["Low", "Average", "Good", "Excellent"]
)
```
